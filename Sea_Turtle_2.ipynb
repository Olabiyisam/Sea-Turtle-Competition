```python
import pandas as pd
import numpy as np
```

```python
dirty = pd.read_csv("dirty_data.csv")
clean = pd.read_csv("cleaned_data.csv")
test = pd.read_csv("test_data.csv")
```

for easy identification of nans and easy handling

```python
clean = pd.DataFrame(np.where(clean.isna(),1,clean),columns=clean.columns)
dirty = pd.DataFrame(np.where(dirty.isna(),1,dirty), columns=clean.columns)
test =  pd.DataFrame(np.where(test.isna(),1,test),columns=clean.columns)
dirty_clean = pd.DataFrame(np.where(dirty.eq(clean),1,0),columns=clean.columns)
```

```python
dirty_clean.drop("Rescue_ID",axis=1,inplace=True)
```

```python
dirty_clean.head()
```

```python
dirty.head()
```

```python
clean.head()
```

```python
clean_len = len(clean)
dirty_len = len(dirty)
test_len = len(test)

print("clean length",clean_len)
print("dirty length", dirty_len)
print("test length", test_len)
```

```python
#concatenate clean, dirty and clean to get the unique value for each column
#this will make it possible to create a one hot encoding

all_data = pd.concat([clean,dirty,test],axis=0)
```

```python
len(all_data)==(clean_len+dirty_len+test_len)
```

```python
column = list(clean.columns)
del column[:2]
```

the main idea is to create an one hot encoding for each column, remember that each field in each column is the target variables,
so we do one hot encoding of the unique features

```python
class_dummies = {}

for col in column:
    
    col_dummy = pd.get_dummies(all_data[col])
    
    class_dummies[col] = col_dummy
```

```python
class_dummies[column[0]].head()
```

```python
train_columns ={}
test_columns = {}
i = 0

for col in column:
    
    train_columns[col] = class_dummies[col].iloc[clean_len:clean_len + dirty_len,:]
    test_columns[col] = class_dummies[col].iloc[clean_len + dirty_len:clean_len + dirty_len+test_len, :]
```

```python
test_columns[column[1]].head()
```

```python
from lightgbm import LGBMClassifier
```

```python
x_train = train_columns[column[0]]
y = dirty_clean[column[0]]
x_test = test_columns[column[0]]
```

```python
model = LGBMClassifier()
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
x,x_val,y,y_val = train_test_split(x_train,y,test_size=0.1,shuffle=True,random_state=42)
```

```python
model.fit(x.values,y)
```

```python
mean_absolute_error(model.predict(x_val),y_val)
```

```python
list(model.predict_proba(x_test))
```
